## Factorial

```python
# 3 --> 3*2*1*0

i = 3
fact = 1 #3
original_i = i
while i>=1:   #3>0 2>0 1>0 0>0
    # print(i)
    # fact *=i
    fact = fact * i   #fact = 1* 3 = 3  3*2 = 6   
    i = i - 1 #3 - 1 = 2 2-1 = 1
print(i)

print(f"Factorial of {original_i}",fact)
```

```python
i = 4
fact = 1
for a in range(i,0,-1):
    fact = fact * a

print(f"factorial of {i} is: ", fact)
```

## Find even and odd numbers

```python
n = int(input("Enter a number: "))
even = []
odd = []
for i in range(1,n):
    if i%2 == 0:
        even.append(i)
    else:
        odd.append(i)

print(f"Even numbers from 1 to {n} is: ", even)
print(f"Odd numbers from 1 to {n} is: ", odd)
```

## Finding the average of 5 numbers using while loop

```python
p = 5                                       #5 - 4 - 3 - 2- 1 
sum = 0                                     #2, 5
count = 0                                   #1, 2

while p > 0:                                #5>0, 4>0
    count += 1                              #count = 1, 2
    f = int(input("Enter the number "))     #1-execute, exe
    sum += f                                #sum = sum + 2 -> 2, 2+3 = 5
    p -= 1                                  # p = p-1(5-1), (4-1)
  
average = sum/count
print("Average of given Numbers:",average)
```
